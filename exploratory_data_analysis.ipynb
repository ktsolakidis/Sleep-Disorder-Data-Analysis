# Exploratory Data Analysis (EDA) - Sleep Health and Lifestyle Dataset

##### We will try to find correlations between variables, and make a story out of our data.

## Key Observations and Insights

### 1. Sleep Duration and Quality of Sleep
We will observe a clear positive relationship between **Sleep Duration** and **Quality of Sleep**. Simply put, longer sleep durations were associated with higher quality sleep. This pattern was especially evident among individuals with **No Disorder**, who had the best sleep quality. On the other hand, those suffering from **Insomnia** experienced lower sleep quality, even if their sleep duration was relatively longer. This could indicate that the nature of sleep (not just its length) plays a critical role in overall sleep quality.

### 2. Physical Activity Level and Daily Steps
The data show that as **Physical Activity Level** increased, so did the number of **Daily Steps**. Individuals who were more active consistently recorded higher step counts, which makes sense—it reflects a physically active lifestyle. This relationship is both intuitive and encouraging, highlighting the importance of daily movement in maintaining a healthy activity level.

### 3. Stress Level and Heart Rate
We notice a moderate positive correlation between **Stress Level** and **Heart Rate**. In simpler terms, higher stress levels often went hand in hand with increased heart rates. Interestingly, this trend was more pronounced in individuals with **Sleep Apnea**, suggesting that their condition might exacerbate the body's physiological response to stress. This highlights the interconnected nature of stress and health conditions.

### 4. Stress Level and Quality of Sleep
Higher **Stress Levels** were linked to reduced **Quality of Sleep**. This finding aligns with common wisdom—stress can take a toll on our sleep. This trend was particularly noticeable among individuals with **Insomnia**, reinforcing how stress can disrupt sleep patterns and quality, creating a vicious cycle.

### 5. Stress Level and Sleep Duration
Unsurprisingly, individuals with higher **Stress Levels** tended to have shorter **Sleep Durations**. This observation is a reminder of the critical impact stress can have on rest and recovery. Addressing stress could therefore be a key step toward improving sleep habits and overall well-being.

### 6. Gender Differences
We observe some differences between genders in sleep patterns. For instance, **Sleep Duration** was slightly longer for females compared to males across all categories of **Sleep Disorder**. On the other hand, **Stress Levels** appeared similar between genders, though males with **Sleep Apnea** showed more variability in their stress responses. These differences suggest the potential for tailored approaches to managing sleep health by gender.

### 7. BMI Category and Lifestyle Indicators
Individuals in the **Obese** and **Overweight** categories reported higher **Stress Levels** compared to those with **Normal Weight**. Additionally, **Overweight** individuals tended to have shorter sleep durations, which could have implications for both their physical and mental health. These findings emphasize the importance of maintaining a healthy BMI for better stress management and sleep quality.

### 8. Sleep Disorder Trends
Those with **No Disorder** generally displayed the best metrics across the board, including **Stress Level**, **Daily Steps**, and **Sleep Duration**. On the other hand, individuals with **Sleep Apnea** and **Insomnia** had higher stress levels and lower sleep quality, highlighting the substantial impact these disorders have on overall lifestyle and health.

### Disclaimer
The following observations and insights are not a substitute for professional medical advice. Individuals experiencing sleep disorders should consult a qualified healthcare provider for accurate diagnosis and tailored treatment plans. However, based on the data, we can say that promoting physical activity, managing stress, and maintaining a healthy BMI are effective strategies that could improve sleep quality and overall health outcomes.


## What We Could Propose


By addressing stress and promoting healthy habits, we have the potential to significantly improve both sleep quality and overall well-being. Encouraging individuals to incorporate physical activity, manage stress through relaxation techniques, and maintain a healthy BMI could create a positive ripple effect on their sleep patterns and overall health. For those with sleep disorders, personalized interventions tailored to their specific challenges—whether stress reduction programs or sleep hygiene improvements—can make a tangible difference. These actionable insights provide a roadmap for healthier lifestyles and better sleep outcomes.


```python
'''
Importing necessary libraries and reading the data.
'''

import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns

df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
```

```python
'''
In the .csv file, when a patient has no sleep disorder, it's written as 'None'.
This causes trouble as pandas and Python understand this as a NaN value.
We need to change this from "None" --> "No Disorder".

We also observe two different values under the "BMI Category", "Normal Weight" and "Normal".
They should be merged into one.

Then we take a look at the first 5 rows of our data.
'''

df.loc[df['Sleep Disorder'].isna(), 'Sleep Disorder'] = 'No Disorder'

df.loc[df['BMI Category'] == 'Normal Weight', 'BMI Category'] = 'Normal'


df.head()
```

```python
df.describe()
```

```python
'''
Let's now try to plot a heatmap out of the numerical values of the data, and try to find some correlations.
'''

sns.heatmap(df.drop(columns=['Person ID']).corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
'''
From the heatmap, we observe some correlated values. 
Quality of Sleep and Sleep Duration,
Daily Steps and Physical Acticity Level,
Stress Level and Heart Rate,
Stress Level and Quality of Sleep,
Stress Level and Sleep Duration.

Let's create some scatter plots to see our observations more analytically.


'''

plt.figure(figsize=(8,3),dpi=300)
sns.regplot(df,x='Sleep Duration',y='Quality of Sleep',line_kws={'color': 'green'})
sns.scatterplot(df,x='Sleep Duration',y='Quality of Sleep',hue='Sleep Disorder')
plt.tight_layout()
plt.show()


plt.figure(figsize=(8,3),dpi=300)
sns.scatterplot(df,x='Daily Steps',y='Physical Activity Level')
sns.regplot(df,x='Daily Steps',y='Physical Activity Level',line_kws={'color': 'green'})
plt.tight_layout()
plt.show()



plt.figure(figsize=(8,3),dpi=300)
sns.scatterplot(df,x='Physical Activity Level',y='Stress Level')
sns.regplot(df,x='Physical Activity Level',y='Stress Level',line_kws={'color': 'green'})
plt.tight_layout()
plt.show()


plt.figure(figsize=(8,3),dpi=300)
sns.regplot(df,x='Stress Level',y='Heart Rate',line_kws={'color': 'green'})
sns.scatterplot(df,x='Stress Level',y='Heart Rate',hue='Sleep Disorder')
plt.tight_layout()
plt.show()


plt.figure(figsize=(8,3),dpi=300)
sns.scatterplot(df,x='Stress Level',y='Quality of Sleep')
sns.regplot(df,x='Stress Level',y='Quality of Sleep',line_kws={'color': 'green'})
plt.tight_layout()
plt.show()


plt.figure(figsize=(8,3),dpi=300)
sns.scatterplot(df,x='Stress Level',y='Sleep Duration')
sns.regplot(df,x='Stress Level',y='Sleep Duration',line_kws={'color': 'green'})
plt.tight_layout()
plt.show()


```
